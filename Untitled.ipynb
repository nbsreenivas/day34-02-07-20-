```
# Importing Libraries
import pandas as pd
import numpy as np
```

```
# Reading the Data:
data = pd.read_csv("https://raw.githubusercontent.com/Mounika-Kajjam/Datasets/master/spam.csv",
                   encoding='latin-1')
data.head()
```

```
data.shape
```

```
# Drop the unnecessary columns
data.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'],axis=1, inplace=True)
data.shape
```

```
# 80% of the data will be in training data and 20% of the data will be in testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.v2, data.v1, 
                                            test_size= 0.2, random_state=1 )
print(X_train.shape)
```

```
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```
##Importing CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer

# creating an object for CountVectorizer
count_vect = CountVectorizer()
```

```
# Generate the word counts for the words in the documents
word_count_vector = count_vect.fit(X_train)

# To get the feature names
word_count_vector.get_feature_names()
```

```
doc_array = word_count_vector.transform(X_train)
doc_array
```

```
# Converting to an array 
doc_array = word_count_vector.transform(X_train).toarray()
doc_array
```

```
# Converting the array into a dataframe
feature_matrix = pd.DataFrame(doc_array,
                              columns= word_count_vector.get_feature_names())
feature_matrix
```
